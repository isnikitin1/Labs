```python
import numpy as np
import matplotlib.pyplot as plt

file = open("D:\Факфиз\Лабы\Механика\Lab3\Mass_Delta_err.txt", 'r')
data = np.genfromtxt(file, delimiter='\t', skip_header=1)

x = data[:, 0]
y = data[:, 1]
x_error = data[:, 2]
y_error = data[:, 3]

#fit1 = np.polyfit(x, y, deg=1)
#p1 = np.poly1d(fit1)

fit2 = np.polyfit(x, y, deg=2)
p2 = np.poly1d(fit2)

plt.figure()
plt.errorbar(x, y, xerr=x_error, yerr=y_error, ecolor='red', fmt='.',  label="Raw data")
plt.plot(x, y, '.')
x = np.linspace(0.1, 1.25, 100)

plt.ylim(5, 13)
#plt.plot(x, p1(x), label='Linear fit')
plt.plot(x, p2(x), label='square fit')
plt.ylabel("Смещение груза, см")
plt.xlabel("Масса, $10^{1}$ г ")

plt.legend()
plt.grid()
plt.title("Экстраполяция зависимости смещения груза от массы пули")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

file = open("D:\Факфиз\Лабы\Механика\Lab3\Mass_Velo_err.txt", 'r')
data = np.genfromtxt(file, delimiter='\t', skip_header=1)

x = data[:, 0]
y = data[:, 1]
y_error = data[:, 2]

fit2 = np.polyfit(x, y, deg=2)
p2 = np.poly1d(fit2)

plt.figure()
plt.errorbar(x, y, yerr=y_error, ecolor='red', fmt='.',  label="Raw data")

plt.plot(y, x, '.')
plt.xlim(0.15,1.2)
plt.ylim(50, 270)
x = np.linspace(0.1, 1.25, 100)
plt.plot(x, p2(x), label='square fit')
plt.ylabel("скорость, м/c")
plt.xlabel("Масса, $10^{1}$ г ")

plt.legend()
plt.grid()
plt.title("Экстраполяция зависимости смещения груза от массы пули")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
file = open("D:\Факфиз\Лабы\Механика\Lab3\Mass_Velo_err.txt", 'r')
data = np.genfromtxt(file, delimiter='\t', skip_header=1)

x = data[:, 0]
y = data[:, 1]
y_error = data[:, 2]
def test_func(x, a):
    return (a/x)**0.5
params, params_covariance = optimize.curve_fit(test_func, x, y, params[0])

plt.figure(figsize=(7,7))
plt.errorbar(x, y,yerr=y_error, ecolor='red', fmt='o',  label="Raw data")


print(params)
x = np.linspace(0.1, 6 , 1000)
plt.plot(x, test_func(x, params[0]), '--', color='green', label='Подогнанная кривая')


plt.xlim(0.1,1.2)
plt.ylabel("Скорость пули, м/c")
plt.xlabel("Масса пули, г ")
plt.legend()
plt.grid()
plt.title("Интерполяция зависимости скорости пули от массы пули")
plt.show()
```
